This firs python notebook will serve as a first data analysis and cleaning. We are looking into the cardiacovascular data and population.

```python
import pandas as pd
from sodapy import Socrata
#aqui sacamos las tablas de datos de hospitales y desfibriladores 
client = Socrata("analisi.transparenciacatalunya.cat", None)

desfibriladors = client.get_all("wpud-ukyg")
df_desfibriladors = pd.DataFrame.from_records(desfibriladors) 

url = "https://www.idescat.cat/serveis/consultes/ca/censph_10_mun_2024.csv"
df_cens = pd.read_csv(url, sep=";", encoding="utf-8")
df_cens.head()
```

```python
path_to_dades = "./dades_mortalitat.xlsx"
df = pd.read_excel(path_to_dades, sheet_name=0, header=None)

print(df[0].unique())
```

```python
comarques = ['ALT CAMP', 'ALT EMPORDÀ', 'ALT PENEDÈS', 'ALT URGELL', 'ALTA RIBAGORÇA',
             'ANOIA', 'BAGES', 'BAIX CAMP', 'BAIX EBRE', 'BAIX EMPORDÀ', 'BAIX LLOBREGAT',
             'BAIX PENEDÈS', 'BARCELONÈS', 'BERGUEDÀ', 'CERDANYA', 'CONCA DE BARBERÀ',
             'GARRAF', 'GARRIGUES', 'GARROTXA', 'GIRONÈS', 'MARESME', 'MOIANÈS', 'MONTSIÀ',
             'NOGUERA', 'OSONA', 'PALLARS JUSSÀ', 'PALLARS SOBIRÀ', "PLA D'URGELL",
             "PLA DE L'ESTANY", 'PRIORAT', "RIBERA D'EBRE", 'RIPOLLÈS', 'SEGARRA', 'SEGRIÀ',
             'SELVA', 'SOLSONÈS', 'TARRAGONÈS', 'TERRA ALTA', 'URGELL',
             'ARAN', 'VALLÈS OCCIDENTAL', 'VALLÈS ORIENTAL']

df['Comarca'] = df[0].where(df[0].isin(comarques))
df['Comarca'] = df['Comarca'].ffill()

sexes = ['Dones', 'Homes', 'Total']
df['Sexe'] = df[0].where(df[0].isin(sexes))
df['Sexe'] = df['Sexe'].ffill()

df.rename(columns={0: 'Cause'}, inplace=True)

cause_labels = [
    "I.Infeccioses", "II.Tumors", "III.Sang", "IV.Endocrines", "V.Mentals",
    "VI.Nerviós", "VII.Ull", "VIII.Oïda", "IX.Circulatori", "X.Respiratori",
    "XI.Digestiu", "XII.Pell", "XIII.Osteomuscular", "XIV.Genitourinari",
    "XV.Embaràs i part", "XVI.Perinatal", "XVII.Congènites",
    "XVIII.Mal definides", "XX.Externes", "XXI. Total"
]

df_final = df[df['Cause'].isin(cause_labels)].copy()

df_final.reset_index(drop=True, inplace=True)

# Now df_final is tidy and ready
print(df_final.head())
```

```python
# Get list of all columns
cols = df_final.columns.tolist()

# Move 'Comarca' and 'Sexe' to the front
cols_ordered = ['Comarca', 'Sexe'] + [c for c in cols if c not in ['Comarca', 'Sexe']]

# Reorder the dataframe
df_final = df_final[cols_ordered]

# Reset index (optional)
df_final.reset_index(drop=True, inplace=True)

# Check result
print(df_final.head())
```

```python
df_final.to_csv("mortality.csv", index=False, encoding="utf-8-sig")
```
